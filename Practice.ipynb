```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
df=pd.read_csv('Salary_Survey.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Dropping columns with no values
df.dropna(axis=1, how='all', inplace=True)
```

```python
# Dropping rows with no values
df.dropna(axis=0, how='all', inplace=True)
```

```python
df.info()
```

```python
new_column_names = [
    "Timestamp", "Age", "Industry", "Job_title", "Additional_job_details",
    "Annual_salary", "Monetary_compensation", "Currency", "Other_currency",
    "Additional_income_details", "Country_of_work", "U.S._state_of_work", "City_of_work",
    "Years_of_experience", "Years_of_experience_in_your_field",
    "Level_of_education", "Gender"
]
```

```python
df.columns = new_column_names
```

```python
df.head()
```

```python
# Dropping columns irrelevant to my analysis
columns_to_drop = [
    "Timestamp", "Additional_job_details", "Monetary_compensation", "Other_currency",
    "Additional_income_details", "U.S._state_of_work", "City_of_work"
]
```

```python
df.drop(columns=columns_to_drop, errors='ignore', inplace=True)
```

```python
df.head()
```

```python
# Cleaning malformed values in Annual_salary column and converting to numerical
df["Annual_salary"] = pd.to_numeric(df["Annual_salary"])
```

```python
# Converting Age column to numerical
df["Age"] = pd.to_numeric(df["Age"], errors='coerce')
```

```python
df.info()
```

```python
# Replacing nan values with zero values
df.fillna(0, inplace=True)
```

```python
df.info()
```

```python
# Converting Currency column into EUR currency
exchange_rates = {
    "USD": 1.0915, "GBP": 0.83358, "EUR": 1.0, "JPY": 164.45, "CAD": 1.4989,
    "AUD": 1.6311, "Other": 1.0, "ZAR": 19.0627, "SEK": 11.4475, "HKD": 8.4110
}
df["Annual_salary_eur"] = df.apply(
    lambda row: row["Annual_salary"] / exchange_rates.get(row["Currency"], 1.0),
    axis=1
)
```

```python
df.drop(columns=["Currency"], inplace=True)
```

```python
# Removing duplicates
df.drop_duplicates(inplace=True)
```

```python
# 10. **Identify and address outliers in `Annual_salary_eur`**
# Using the Interquartile Range (IQR) method
Q1 = df["Annual_salary_eur"].quantile(0.25)
Q3 = df["Annual_salary_eur"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
# Filter to only keep rows within the acceptable range
df = df[(df["Annual_salary_eur"] >= lower_bound) & (df["Annual_salary_eur"] <= upper_bound)]
```

```python
# 11. **Feature Selection**: Choose relevant features (example selection here)
selected_features = [
    "Age", "Industry", "Job_title", "Annual_salary_eur",
    "Country_of_work", "Years_of_experience", "Years_of_experience_in_your_field",
    "Level_of_education", "Gender"
]
```

```python
df = df[selected_features]
```
